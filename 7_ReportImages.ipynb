```python
from PIL import Image
import matplotlib.pyplot as plt

# List of selected image files
selected_files = [
    "generated_images_GAN/epoch_16_image_6.png",
    "generated_images_GAN/epoch_24_image_3.png",
    "generated_images_GAN/epoch_30_image_50.png",
    "generated_images_GAN/epoch_31_image_22.png",
    "generated_images_GAN/epoch_36_image_22.png",
    "generated_images_GAN/epoch_43_image_62.png",
    "generated_images_GAN/epoch_50_image_57.png",
    "generated_images_GAN/epoch_52_image_36.png",
    "generated_images_GAN/epoch_62_image_57.png",
    "generated_images_GAN/epoch_62_image_50.png"
]

# Create a blank canvas for the grid
grid_size = (5, 2)  # Rows, Columns
cell_margin = 8  # Margin between cells
canvas_width = grid_size[1] * 64 + (grid_size[1] - 1) * cell_margin  # Total canvas width
canvas_height = grid_size[0] * 64 + (grid_size[0] - 1) * cell_margin  # Total canvas height

# Create the canvas with a white background
canvas = Image.new("RGB", (canvas_width, canvas_height), "white")

# Iterate over the selected files and place them in the grid
for index, file_path in enumerate(selected_files):
    row = index // grid_size[1]
    col = index % grid_size[1]

    # Open the image
    image = Image.open(file_path)

    # Calculate the position to paste the image on the canvas
    x = col * (64 + cell_margin)
    y = row * (64 + cell_margin)

    # Paste the image onto the canvas
    canvas.paste(image, (x, y))


grid_image_path = "grid_image.jpg"
canvas.save(grid_image_path)
    
# Display the grid using matplotlib
plt.imshow(canvas)
plt.axis('off')
plt.show()
```
